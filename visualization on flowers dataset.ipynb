```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('flowers.csv')
```

# **Boxplot**

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=dataset, x="species", y="sepal length")
plt.title("Boxplot of Sepal Length by Species")
plt.show()
```

# **Pairplot**

```python
sns.pairplot(dataset, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

# **Violinplot**

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x="species", y="sepal length", data=dataset, palette="muted")
plt.title("Violin Plot of Sepal Length by Species")
plt.show()
```

# **Scatterplot**

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x="sepal length", y="sepal width", hue="species", data=dataset, s=70) # Removed extra spaces from column names
plt.title("2D Scatter Plot of Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.show()
```

# **Piechart**

```python
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 2)
species_counts = dataset['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Proportion of Species in the Dataset')
plt.tight_layout()
plt.legend(title="Species")
plt.show()
```

# **Histograph**

```python
plt.figure(figsize=(12, 8))
plt.subplot(1, 1, 1)
plt.hist(dataset['species'], bins=10, color='blue')
plt.xlabel('species')
plt.ylabel('quantity')
plt.title('Histogram of species')
```

# **Bargraph**

```python
target = dataset['species']
target_names = ['setosa', 'versicolor', 'virginica']
species_counts = pd.Series(target).value_counts().sort_index()
plt.figure(figsize=(12, 8))
plt.bar(target_names, species_counts, color=['lightblue', 'lightgreen', 'lightcoral'])
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Count of Each Species in the Iris Dataset")
plt.show()
```
